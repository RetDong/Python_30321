```python
a=np.arange(-10,10,0.5)
a
```

```python
b=a*a
a
```

```python
c=a*10
plt.plot(a,b)
plt.plot(a,c)
```

```python
t=np.arange(0,12,1)
y1=np.sin(t)
y2=np.cos(t)

plt.plot(t,y1,'r--',marker='*')
plt.plot(t,y2,'g--')

plt.grid()
plt.figure(figsize=(12,2))
```
